# In this code we merge the two main datasets we have, the one with information of each well but without time related (such as slope, drainage and more), with the one with time data reletead (such as precipitation, NDVI and more)

```python
import pandas as pd
```

```python
# Read the CSV file
wells_df = pd.read_csv('../../data/processed_data/igrac/wells_gambia_final.csv')
wells_time_df = pd.read_csv('../../data/processed_data/final_data_gambia.csv')
```

```python
wells_df.head()
```

```python
wells_time_df.head()
```

```python
# Merging the dataframes on the 'ID' column
merged_df = pd.merge(left=wells_time_df, right=wells_df[['ID', 'DepthToGroundwater', 
                                                         'Curvatu_tif2', 'Drainage_density', 
                                                         'Slope_tif2', 'Hydrogeo', 'NASADEM_HGT']], 
                     on='ID', 
                     how='left')
merged_df
```

```python
# Saving the merged dataframe to a new CSV file (optional)
merged_df.to_csv('../../data/processed_data/wells_data_gambia_for_machine_learning.csv', index=False)
```
